```python
import seaborn as sns
import pandas
import matplotlib as mat
import numpy
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

data = pandas.read_csv("../Data/sales_predictions.csv")
```

```python
data.info()
```

```python
sns.relplot(x='date',y='item_price',data=data)
```

```python
data50=data.sample(50)
sns.relplot(x='date',y='item_price',data=data50)
```

```python
plt.figure(figsize(20,6))
sns.relplot(x='date',y='item_price',data=data50)
```

```python
f, ax = plt.subplots(figsize=(20,6))
sns.relplot(x="date", y="item_price", data = data50, kind="line", ax = ax)
plt.figure(figsize=(20,6))
plt.show()
```

```python
pandas.to_datetime(data50.date)
```

```python
f, ax = plt.subplots(figsize=(20,6))
data50.date=pandas.to_datetime(data50.date)
sns.relplot(x="date", y="item_price", data = data50, kind="line", ax = ax)
plt.show()
```
